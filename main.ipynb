```python
#get your current location
import geocoder
from geopy.geocoders import ArcGIS
import openrouteservice as ors
import folium

client = ors.Client(key = '5b3ce3597851110001cf6248a1038de5c5304bb39fdff93c8db3113d')

m = folium.Map(locations = list([10, 85]), tiles = "cartodbpositron", zoom_start = 14)

def get_current_coordinates():
    g = geocoder.ip('me')
    print("Current location : ", g.latlng)
    return g.latlng

#destination coordinates
def dest_location():
    nom = ArcGIS()
    dest = input('Enter  the location : ')
    dest_coord = list(nom.geocode(dest))
    dest_coord = list(dest_coord[1])
    print("Final location : ", dest_coord)
    return dest_coord

def swap(b = []):
    a = b[:]
    a[0], a[1] = a[1], a[0]
    return a

#route between current and destination
def finalroute(s, e):
    coords = [s, e]
    route = client.directions(coordinates = coords,
                             profile = 'driving-car',
                             format = 'geojson')
    print("\n Route Coordinates : ")
    for coord in route['features'][0]['geometry']['coordinates']:
        print(coord)
    return route['features'][0]['geometry']['coordinates']
    #for coord in route['features'][0]['geometry']['coordinates']:
        #folium.PolyLine(locations = , color = 'blue').add_to(m)

start = get_current_coordinates()
start_s = swap(start)
end = dest_location()
end_s = swap(end)

direction = finalroute(start_s, end_s)

folium.Marker(location = start).add_to(m)
folium.Marker(location = end).add_to(m)
```
