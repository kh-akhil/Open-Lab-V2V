### Get current co-ordinates, destination co-ordinates, and route between these 2

```python
#get your current location
import geocoder
from geopy.geocoders import ArcGIS
import openrouteservice as ors
import folium
import operator
from functools import reduce

client = ors.Client(key = '5b3ce3597851110001cf6248a1038de5c5304bb39fdff93c8db3113d')

m = folium.Map(locations = list([80, 10]), tiles = "cartodbpositron", zoom_start = 14)

def get_current_coordinates():
    g = geocoder.ip('me')
    print("Current location : ", g.latlng)
    return g.latlng

#destination coordinates
def dest_location():
    nom = ArcGIS()
    dest = input('Enter  the location : ')
    dest_coord = list(nom.geocode(dest))
    dest_coord = list(dest_coord[1])
    print("Final location : ", dest_coord)
    return dest_coord

def swap(b = []):
    a = b[:]
    a[0], a[1] = a[1], a[0]
    return a

#route between current and destination
def finalroute(s, e):
    coords = [s, e]
    route = client.directions(coordinates = coords,
                             profile = 'driving-car',
                             format = 'geojson')
    folium.PolyLine(locations=[list(reversed(coord)) for coord in route['features'][0]['geometry']['coordinates']], color="blue").add_to(m)
    print("Route Coordinates : ")
    for coord in route['features'][0]['geometry']['coordinates']:
        print(coord)
    return route['features'][0]['geometry']['coordinates']

#start = get_current_coordinates()
start = [12.895033, 77.675858]
print("Current location : ", start)
start_s = swap(start)
end = dest_location()
end_s = swap(end)

direction = finalroute(start_s, end_s)

folium.Marker(location = start).add_to(m)
folium.Marker(location = end).add_to(m)
```

### Detect if car is in the route

```python
#define car positions 
car_positions = {
    1: [77.5946, 12.9716],
    2: [77.5913, 12.9791],
    3: [77.6213, 12.9200],
    4: [77.6271, 12.9279],
    5: [77.7018, 12.9561],
    6: [77.630917, 13.026247],
    7: [77.6169, 12.9044],
    8: [77.6245, 12.9352],
    9: [77.5847, 12.9528],
    10: [77.65297, 13.029655],
    11: [77.748162, 12.899122],
    12: [77.781409, 12.856742],
    13: [77.725157, 12.767398],
    14: [72.878057, 21.204618],
    15: [77.717389, 13.256997]
}
count = 0
for coord in direction:
    for key, value in car_positions.items():
        if value==coord:
            count+=1
            print(value)
            print("Car ID ", key, "found in the selected route")
            print("Send Alert Message")
            folium.Marker(location = list(swap(value)), color = "red").add_to(m)

print(count, "vehicles found in the route")
```
